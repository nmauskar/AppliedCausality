## Gibbs Sampler

```julia
mutable struct gibbsSampler
    A::Int64
    I::Int64
    T::Int64
    K::Int64
    I_K::Array{Float64, 2}
    σU::Float64
    σV::Float64
    σW::Float64
    σY::Float64
    Y::Array{Float64, 3}
    U::Array{Float64, 2}
    V::Array{Float64, 2}
    W::Array{Float64, 2}
end

function initGibbs(A::Int64, I::Int64, T::Int64, K::Int64, σU::Float64, σV::Float64, σW::Float64, σY::Float64)
    Y = zeros(A, I, T)
    U = zeros(I, K)
    V = zeros(T, K)
    W = zeros(A, K)
    I_K = diagm(ones(K))
    return gibbsSampler(A, I, T, K, I_K, σU, σV, σW, σY, Y, U, V, W)
end
```

```julia
function sampleU!(gs::gibbsSampler)
    for k ∈ 1:gs.K
        for i ∈ 1:gs.I
            gs.U[i,k] = rand(Normal(0.0, gs.σU))
        end
    end
    return gs.U
end

function sampleV!(gs::gibbsSampler)
    for k ∈ 1:gs.K
        for t ∈ 1:gs.T
            gs.V[t,k] = rand(Normal(0.0, gs.σV))
        end
    end
    return gs.V
end

function sampleW!(gs::gibbsSampler)
    for k ∈ 1:gs.K
        for a ∈ 1:gs.A
            gs.W[a,k] = rand(Normal(0.0, gs.σW))
        end
    end
    return gs.W
end

function sampleY!(gs::gibbsSampler)
    for t ∈ 1:gs.T
        for i ∈ 1:gs.I
            for a ∈ 1:gs.A
                gs.Y[a, i, t] = rand(Normal(sum(gs.U[i, :] .* gs.V[t, :] .* gs.W[a, :]), gs.σY))
            end
        end
    end
    return gs.Y
end

function samplePrior!(gs::gibbsSampler)
    sampleU!(gs)
    sampleV!(gs)
    sampleW!(gs)
    sampleY!(gs)
end
```

```julia
function updateU!(gs::gibbsSampler)
    S = zeros(gs.K, gs.K)
    b = zeros(gs.K)
    
    for t ∈ 1:gs.T
        for a ∈ 1:gs.A
            S += (gs.V[t, :] .* gs.W[a, :]) * (gs.V[t, :] .* gs.W[a, :])'
        end
    end
    
    for i ∈ 1:gs.I
        b .= 0.0
        for t ∈ 1:gs.T
            for a ∈ 1:gs.A
                b += (gs.V[t, :] .* gs.W[a, :]) .* gs.Y[a,i,t]
            end
        end
        MVN = MvNormalCanon(b/gs.σY, 1.0/gs.σU .* gs.I_K + S./gs.σY)
        gs.U[i, :] = rand(MVN)
    end
    return gs.U
end

function updateV!(gs::gibbsSampler)
    S = zeros(gs.K, gs.K)
    b = zeros(gs.K)
    
    for i ∈ 1:gs.I
        for a ∈ 1:gs.A
            S += (gs.U[i, :] .* gs.W[a, :]) * (gs.U[i, :] .* gs.W[a, :])'
        end
    end
    
    for t ∈ 1:gs.T
        b .= 0.0
        for i ∈ 1:gs.I
            for a ∈ 1:gs.A
                b += (gs.U[i, :] .* gs.W[a, :]) .* gs.Y[a,i,t]
            end
        end
        MVN = MvNormalCanon(b/gs.σY, 1.0/gs.σV .* gs.I_K + S./gs.σY)
        gs.V[t, :] = rand(MVN)
    end
    return gs.V
end

function updateW!(gs::gibbsSampler)
    S = zeros(gs.K, gs.K)
    b = zeros(gs.K)
    
    for i ∈ 1:gs.I
        for t ∈ 1:gs.T
            S += (gs.U[i, :] .* gs.V[t, :]) * (gs.U[i, :] .* gs.V[t, :])'
        end
    end
    
    for a ∈ 1:gs.A
        b .= 0.0
        for i ∈ 1:gs.I
            for t ∈ 1:gs.T
                b += (gs.U[i, :] .* gs.V[t, :]) .* gs.Y[a,i,t]
            end
        end
        MVN = MvNormalCanon(b/gs.σY, 1.0/gs.σW .* gs.I_K + S./gs.σY)
        gs.W[a, :] = rand(MVN)
    end
    return gs.W
end

function updateY!(gs::gibbsSampler, mask::Set{Tuple{Int64, Int64, Int64}})
    for i ∈ mask
        gs.Y[i[1], i[2], i[3]] = rand(Normal(sum(gs.U[i[2], :] .* gs.V[i[3], :] .* gs.W[i[1], :]), gs.σY))
    end
    return gs.Y
end

function updateGibbs!(gs::gibbsSampler, mask::Set{Tuple{Int64, Int64, Int64}})
    updateU!(gs)
    updateW!(gs)
    updateV!(gs)
    updateY!(gs, mask)
    return gs
end
```

## Geweke Testing

```julia
A = 2
I = 2
T = 2
K = 2
σ = 1.0

numItr = 100000

gs = initGibbs(A, I, T, K, σ, σ, σ, σ)

statU = zeros(2, numItr)
statV = zeros(2, numItr)
statW = zeros(2, numItr)
statY = zeros(2, numItr)
mask = Set{Tuple{Int64, Int64, Int64}}()

for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            push!(mask, (a, i, t))
        end
    end
end

@showprogress for i ∈ 1:numItr
    samplePrior!(gs)
    statU[1, i] = mean(gs.U)
    statV[1, i] = mean(gs.V)
    statW[1, i] = mean(gs.W)
    statY[1, i] = mean(gs.Y)
end
@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end
@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    statU[2, i] = mean(gs.U)
    statV[2, i] = mean(gs.V)
    statW[2, i] = mean(gs.W)
    statY[2, i] = mean(gs.Y)
end

display(qqplot(statU[1,:], statU[2,:], title="U", fmt=:png))
display(qqplot(statV[1,:], statV[2,:], title="V", fmt=:png))
display(qqplot(statW[1,:], statW[2,:], title="W", fmt=:png))
display(qqplot(statY[1,:], statY[2,:], title="Y", fmt=:png))
```

## California Smoking Data

```julia
data = load("smoking.rda")["smoking"]
numStates = 39
numYears = 31
# California is state 3 and the intervention happened in 1989
A = 2
T = numYears
I = numStates

YObs = zeros(A, I, T)
for row ∈ eachrow(data)
    if row["state"] == 3 && Int64(row["year"]) > 1988
        YObs[2, Int64(row["state"]), Int64(row["year"] - 1969.0)] = row["cigsale"] 
    else
        YObs[1, Int64(row["state"]), Int64(row["year"] - 1969.0)] = row["cigsale"] 
    end
end

mask = Set{Tuple{Int64, Int64, Int64}}()

for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            if YObs[a,i,t] == 0.0
                push!(mask, (a, i, t))
            end
        end
    end
end
```

```julia
e = eigen(YObs[1, :, 1:1989-1969+1]*YObs[1,:,1:1989-1969+1]')
bar(real.(e.values))
```

```julia
numItr = 1000
K = 1
σ = 1.0

gs = initGibbs(A, I, T, K, σ, σ, σ, σ)

samplePrior!(gs)
gs.Y = deepcopy(YObs)

sampleY = zeros(numItr, A, I, T)

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    sampleY[i, :, :, :] = deepcopy(gs.Y)
end
```

```julia
#Standard California Synthetic Control
df = load_smoking()
panel = load_smoking_panel()
s = SimpleSCM(panel)
SynthControl.fit!(s)
```

```julia
plot([YObs[1, 3, 1:19]; YObs[2, 3, 20:end]], label="Observed")
plot!(mean(sampleY[:, 1, 3, :], dims=1)', label="Bayesian Synthetic")
plot!(s.ŷ₁, label="Abadie Synthetic")
plot!([19], label="Intervention",  seriestype = :vline, line=:dash)
```

```julia
numItr = 1000
K = 1
σ = 1.0

gs = initGibbs(A, I, T, K, σ, σ, σ, σ)

samplePrior!(gs)
gs.Y = deepcopy(YObs)

sampleY = zeros(numItr, A, I, T)

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    sampleY[i, :, :, :] = deepcopy(gs.Y)
end
```

## Brexit Data

```julia
#Standard Brexit Synthetic Control, Intervention happeed at timestep 31, Britain is country 22
df = load_brexit()
A = 2
I = 23
T = 39

YObs = zeros(A, I, T)

IMap = unique(df[!, "country"])
TMap = unique(df[!, "quarter"])

for i ∈ eachrow(df)
    state = findfirst(==(i["country"]), IMap)
    time = findfirst(==(i["quarter"]), TMap)
    if state == 22 && time > 30
        YObs[2, state, time] = i["realgdp"]
    else
        YObs[1, state, time] = i["realgdp"]
    end
end

mask = Set{Tuple{Int64, Int64, Int64}}()

for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            if YObs[a,i,t] == 0.0
                push!(mask, (a, i, t))
            end
        end
    end
end
```

```julia
e = eigen(YObs[1, :, 1:30]*YObs[1,:,1:30]')
bar(real.(log.(Complex.(e.values))))
bar(real.(e.values))
```

```julia
panel = load_brexit_panel()
s = SimpleSCM(panel)
SynthControl.fit!(s)
```

```julia
numItr = 1000
K = 4
σ = 1.0

gs = initGibbs(A, I, T, K, σ, σ, σ, σ)

samplePrior!(gs)
gs.Y = deepcopy(YObs)

sampleY = zeros(numItr, A, I, T)

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    sampleY[i, :, :, :] = deepcopy(gs.Y)
end
```

```julia
plot([YObs[1, 22, 1:30]; YObs[2, 22, 31:end]], label="Observed")
plot!(mean(sampleY[:, 1, 22, :], dims=1)', label="Bayesian Synthetic")
plot!(s.ŷ₁, label="Abadie Synthetic")
plot!([30], label="Intervention",  seriestype = :vline, line=:dash)
```
