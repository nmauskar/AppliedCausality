```julia
function SoftmaxLSE(x::Array{Float64})
    c = maximum(x)
    LSE = c + log(sum(exp.(x .- c)))
    return exp.(x .- LSE)
end
```

## Gibbs Sampler

```julia
mutable struct gibbsSampler
    A::Int64
    I::Int64
    T::Int64
    K::Int64
    I_K::Array{Float64, 2}
    σU::Float64
    σV::Float64
    σW::Float64
    σX::Float64
    σY::Float64
    α::Array{Float64}
    Y::Array{Float64, 2}
    X::Array{Float64, 3}
    U::Array{Float64, 2}
    V::Array{Float64, 2}
    W::Array{Float64, 2}
    D::Array{Float64, 2}
    C::Array{Float64}
end

function initGibbs(A::Int64, I::Int64, T::Int64, K::Int64, σU::Float64, σV::Float64, σW::Float64, σX::Float64, σY::Float64, α::Array{Float64})
    Y = zeros(A, T)
    X = zeros(A, I, T)
    D = zeros(A, I)
    C = zeros(A)
    U = zeros(I, K)
    V = zeros(T, K)
    W = zeros(A, K)
    I_K = diagm(ones(K))
    return gibbsSampler(A, I, T, K, I_K, σU, σV, σW, σX, σY, α, Y, X, U, V, W, D, C)
end
```

```julia
function sampleU!(gs::gibbsSampler)
    gs.U = rand(Normal(0.0, gs.σU), gs.I, gs.K)
    return gs.U
end

function sampleV!(gs::gibbsSampler)
    gs.V = rand(Normal(0.0, gs.σV), gs.T, gs.K)
    return gs.V
end

function sampleW!(gs::gibbsSampler)
    gs.W = rand(Normal(0.0, gs.σW), gs.A, gs.K)
    return gs.W
end

function sampleX!(gs::gibbsSampler)
    for t ∈ 1:gs.T
        for i ∈ 1:gs.I
            for a ∈ 1:gs.A
                gs.X[a, i, t] = rand(Normal(sum(gs.U[i, :] .* gs.V[t, :] .* gs.W[a, :]), gs.σX))
            end
        end
    end
end

function sampleD!(gs::gibbsSampler)
    gs.D = rand(Dirichlet(gs.α), gs.A)'
    return gs.D
end

function sampleC!(gs::gibbsSampler)
    for a ∈ 1:gs.A
        gs.C[a] = rand(Categorical(gs.D[a, :]))
    end
    return gs.C
end

function sampleY!(gs::gibbsSampler)
    for t ∈ 1:gs.T
        for a ∈ 1:gs.A
            gs.Y[a, t] = rand(Normal(sum(gs.U[Int64(gs.C[a]), :] .* gs.V[t, :] .* gs.W[a, :]), gs.σY))
        end
    end
    return gs.Y
end

function samplePrior!(gs::gibbsSampler)
    sampleU!(gs)
    sampleV!(gs)
    sampleW!(gs)
    sampleX!(gs)
    sampleD!(gs)
    sampleC!(gs)
    sampleY!(gs)
end
```

```julia
A = 2
I = 50
T = 30
K = 3
σ = 1.0
α = ones(I)

gs = initGibbs(A, I, T, K, σ, σ, σ, σ, σ, α)

@time for i ∈ 1:1000 samplePrior!(gs) end
```

```julia
function updateU!(gs::gibbsSampler)
    SX = zeros(gs.K, gs.K)
    SY = zeros(gs.K, gs.K)

    bY = zeros(gs.K)
    bX = zeros(gs.K)
    
    for t ∈ 1:gs.T
        for a ∈ 1:gs.A
            SX += (gs.V[t, :] .* gs.W[a, :]) * (gs.V[t, :] .* gs.W[a, :])'
        end
    end
    
    for i ∈ 1:gs.I
        bX .= 0.0
        bY .= 0.0
        SY .= 0.0
        
        for t ∈ 1:gs.T
            for a ∈ 1:gs.A
                bX += (gs.V[t, :] .* gs.W[a, :]) .* gs.X[a,i,t]
                if i == Int64(gs.C[a])
                    bY += (gs.V[t, :] .* gs.W[a, :]) .* gs.Y[a,t]
                    SY += (gs.V[t, :] .* gs.W[a, :]) * (gs.V[t, :] .* gs.W[a, :])'
                end
            end
        end
        
        MVN = MvNormalCanon(bY./gs.σY + bX./gs.σX, 1.0/gs.σU .* gs.I_K + SX./gs.σX + SY./gs.σY)
        gs.U[i, :] = rand(MVN)
    end
    return gs.U
end

function updateV!(gs::gibbsSampler)
    SX = zeros(gs.K, gs.K)
    SY = zeros(gs.K, gs.K)

    bY = zeros(gs.K)
    bX = zeros(gs.K)
    
    for i ∈ 1:gs.I
        for a ∈ 1:gs.A
            SX += (gs.U[i, :] .* gs.W[a, :]) * (gs.U[i, :] .* gs.W[a, :])'
            if i == Int64(gs.C[a])
                SY += (gs.U[i, :] .* gs.W[a, :]) * (gs.U[i, :] .* gs.W[a, :])'
            end
        end
    end
    
    for t ∈ 1:gs.T
        bX .= 0.0
        bY .= 0.0
        for i ∈ 1:gs.I
            for a ∈ 1:gs.A
                bX += (gs.U[i, :] .* gs.W[a, :]) .* gs.X[a,i,t]
                if i == Int64(gs.C[a])
                    bY += (gs.U[i, :] .* gs.W[a, :]) .* gs.Y[a,t]
                end
            end
        end
        MVN = MvNormalCanon(bY./gs.σY + bX./gs.σX, 1.0/gs.σV .* gs.I_K + SY./gs.σY + SX./gs.σX)
        gs.V[t, :] = rand(MVN)
    end
    return gs.V
end

function updateW!(gs::gibbsSampler)
    SX = zeros(gs.K, gs.K)
    SY = zeros(gs.K, gs.K)

    bY = zeros(gs.K)
    bX = zeros(gs.K)
    
    for i ∈ 1:gs.I
        for t ∈ 1:gs.T
            SX += (gs.U[i, :] .* gs.V[t, :]) * (gs.U[i, :] .* gs.V[t, :])'
        end
    end
    
    for a ∈ 1:gs.A
        bX .= 0.0
        bY .= 0.0
        SY .= 0.0
        for i ∈ 1:gs.I
            for t ∈ 1:gs.T
                bX += (gs.U[i, :] .* gs.V[t, :]) .* gs.X[a,i,t]
                if i == Int64(gs.C[a])
                    bY += (gs.V[t, :] .* gs.W[a, :]) .* gs.Y[a,t]
                    SY += (gs.V[t, :] .* gs.W[a, :]) * (gs.V[t, :] .* gs.W[a, :])'
                end
            end
        end
        MVN = MvNormalCanon(bY./gs.σY + bX./gs.σX, 1.0/gs.σW .* gs.I_K + SY./gs.σY + SX./gs.σX)
        gs.W[a, :] = rand(MVN)
    end
    return gs.W
end

function updateD!(gs::gibbsSampler)
    for a ∈ 1:gs.A
        param = deepcopy(gs.α)
        param[Int64(gs.C[a])] += 1.0
        gs.D[a, :] = rand(Dirichlet(param))
    end
    return gs.D
end

function updateC!(gs::gibbsSampler)
    for a ∈ 1:gs.A
        probs = zeros(gs.I)
        for i ∈ 1:gs.I
            N = Float64(gs.T)/2.0 * log(2*π) - gs.T*log(gs.σY) - 0.5 * 1.0/gs.σY * (gs.Y[a, :] - gs.X[a, Int64(i), :])' * (gs.Y[a, :] - gs.X[a, Int64(i), :])
            probs[i] = N - log(gs.D[i] + 1e-8)
        end
        newProbs = SoftmaxLSE(probs)
        gs.C[a] = rand(Categorical(newProbs ./sum(newProbs)))
    end
    return gs.C
end

function updateY!(gs::gibbsSampler, mask::Set{Tuple{Int64, Int64}})
    for i ∈ mask
        gs.Y[i[1], i[2]] = rand(Normal(sum(gs.U[Int64(gs.C[i[1]]), :] .* gs.V[i[2], :] .* gs.W[i[1], :]), gs.σY))
    end
    return gs.Y
end

function updateGibbs!(gs::gibbsSampler, mask::Set{Tuple{Int64, Int64}})
    updateU!(gs)
    updateW!(gs)
    updateV!(gs)
    updateC!(gs)
    updateD!(gs)
    updateY!(gs, mask)
    return gs
end

function updateGeweke!(gs::gibbsSampler, mask::Set{Tuple{Int64, Int64}})
    updateU!(gs)
    updateW!(gs)
    updateV!(gs)
    updateC!(gs)
    updateD!(gs)
    sampleX!(gs)
    updateY!(gs, mask)
    return gs
end
```

## Geweke Testing

```julia
A = 2
I = 2
T = 2
K = 2
σ = 1.0
α = ones(I)


numItr = 10000

gs = initGibbs(A, I, T, K, σ, σ, σ, σ, σ, α)

statU = zeros(2, numItr)
statV = zeros(2, numItr)
statW = zeros(2, numItr)
statY = zeros(2, numItr)
statD = zeros(2, numItr)
statC = zeros(2, numItr)
statX = zeros(2, numItr)

mask = Set{Tuple{Int64, Int64}}()

for a ∈ 1:A
    for t ∈ 1:T
        push!(mask, (a, t))
    end
end

@showprogress for i ∈ 1:numItr
    samplePrior!(gs)
    statU[1, i] = mean(gs.U)
    statV[1, i] = mean(gs.V)
    statW[1, i] = mean(gs.W)
    statY[1, i] = mean(gs.Y)
    statD[1, i] = maximum(gs.D)
    statC[1, i] = mean(gs.C)
    statX[1, i] = mean(gs.X)
end
@showprogress for i ∈ 1:numItr
    updateGeweke!(gs, mask)
end
@showprogress for i ∈ 1:numItr
    updateGeweke!(gs, mask)
    statU[2, i] = mean(gs.U)
    statV[2, i] = mean(gs.V)
    statW[2, i] = mean(gs.W)
    statY[2, i] = mean(gs.Y)
    statD[2, i] = maximum(gs.D)
    statC[2, i] = mean(gs.C)
    statX[2, i] = mean(gs.X)
end

display(qqplot(statU[1,:], statU[2,:], title="U", fmt=:png))
display(qqplot(statV[1,:], statV[2,:], title="V", fmt=:png))
display(qqplot(statW[1,:], statW[2,:], title="W", fmt=:png))
display(qqplot(statY[1,:], statY[2,:], title="Y", fmt=:png))
display(qqplot(statD[1,:], statD[2,:], title="D", fmt=:png))
display(qqplot(statC[1,:], statC[2,:], title="C", fmt=:png))
display(qqplot(statX[1,:], statX[2,:], title="X", fmt=:png))
```

## California Smoking Data

```julia
data = load("smoking.rda")["smoking"]
numStates = 39
numYears = 31
# California is state 3 and the intervention happened in 1989
A = 2
T = numYears
I = numStates

Y = zeros(A, T)
X = zeros(A, I-1, T)

YObs = zeros(A, I, T)
for row ∈ eachrow(data)
    if row["state"] == 3 && Int64(row["year"]) > 1988
        YObs[2, Int64(row["state"]), Int64(row["year"] - 1969.0)] = row["cigsale"]
    else
        YObs[1, Int64(row["state"]), Int64(row["year"] - 1969.0)] = row["cigsale"] 
    end
end

Y = deepcopy(YObs[:, 3, :])
for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            if i > 3
                X[a, i-1, t] = YObs[a,i,t]
            elseif i < 3
                X[a, i, t] = YObs[a,i,t]
            end
        end
    end
end

mask = Set{Tuple{Int64, Int64}}()

for a ∈ 1:A
    for t ∈ 1:T
        if Y[a,t] == 0.0
            push!(mask, (a, t))
        end
    end
end
```

```julia
e = eigen(YObs[1, :, 1:1989-1969+1]*YObs[1,:,1:1989-1969+1]')
bar(real.(e.values))
```

```julia
numItr = 1000
K = 1
σ = mean(X)
α = 1.0/(I-1) * ones(I-1)
gs = initGibbs(A, I-1, T, K, σ, σ, σ, σ, σ, α)

samplePrior!(gs)
gs.Y = deepcopy(Y)
gs.X = deepcopy(X)

sampleY = zeros(numItr, A, T)
sampleC = zeros(numItr, A)

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    sampleY[i, :, :] = deepcopy(gs.Y)
    sampleC[i, :] = deepcopy(gs.C)
end
```

```julia
#Standard California Synthetic Control
df = load_smoking()
panel = load_smoking_panel()
s = SimpleSCM(panel)
SynthControl.fit!(s)
```

```julia
histogram(sampleC[:, 1], bins=I-1)
```

```julia
plot([YObs[1, 3, 1:19]; YObs[2, 3, 20:end]], label="Observed")
plot!(mean(sampleY[:, 1, :], dims=1)', label="Bayesian Synthetic")
#plot!(sampleY[end-5, 1, :], label="Single sample Bayesian Synthetic")

plot!(s.ŷ₁, label="Abadie Synthetic")
plot!([19], label="Intervention",  seriestype = :vline, line=:dash)
```

## Germany Data

```julia
# Germany is region 7 and the intervention happened in 1990 or timestep 30
df = load_germany()

A = 2
I = 18
T = 44
Y = zeros(A, T)
X = zeros(A, I-1, T)

YObs = zeros(A, I, T)

for i ∈ eachrow(df)
    if i["index"] == 12
        i["index"] = 11
    end
    if i["index"] == 14
        i["index"] = 12
    end
    if i["index"] == 16
        i["index"] = 13
    end
    if i["index"] == 18
        i["index"] = 14
    end
    if i["index"] == 19
        i["index"] = 15
    end
    if i["index"] == 20
        i["index"] = 16
    end
    if i["index"] == 21
        i["index"] = 17
    end
    if i["index"] == 7 && i["year"] > 1989
        YObs[2, i["index"], i["year"]-1960+1] = i["gdp"]
    else
        YObs[1, i["index"], i["year"]-1960+1] = i["gdp"]
    end
end

Y = deepcopy(YObs[:, 7, :])
for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            if i > 7
                X[a, i-1, t] = YObs[a,i,t]
            elseif i < 7
                X[a, i, t] = YObs[a,i,t]
            end
        end
    end
end

mask = Set{Tuple{Int64, Int64}}()

for a ∈ 1:A
    for t ∈ 1:T
        if Y[a,t] == 0.0
            push!(mask, (a, t))
        end
    end
end
```

## Brexit Data

```julia
#Standard Brexit Synthetic Control, Intervention happeed at timestep 31, Britain is country 22
df = load_brexit()
A = 2
I = 23
T = 39

Y = zeros(A, T)
X = zeros(A, I-1, T)

YObs = zeros(A, I, T)

IMap = unique(df[!, "country"])
TMap = unique(df[!, "quarter"])

for i ∈ eachrow(df)
    state = findfirst(==(i["country"]), IMap)
    time = findfirst(==(i["quarter"]), TMap)
    if state == 22 && time > 30
        YObs[2, state, time] = i["realgdp"]
    else
        YObs[1, state, time] = i["realgdp"]
    end
end


Y = deepcopy(YObs[:, 22, :])
for a ∈ 1:A
    for i ∈ 1:I
        for t ∈ 1:T
            if i > 22
                X[a, i-1, t] = YObs[a,i,t]
            elseif i < 22
                X[a, i, t] = YObs[a,i,t]
            end
        end
    end
end

mask = Set{Tuple{Int64, Int64}}()

for a ∈ 1:A
    for t ∈ 1:T
        if Y[a,t] == 0.0
            push!(mask, (a, t))
        end
    end
end
```

```julia
e = eigen(YObs[1, :, 1:30]*YObs[1,:,1:30]')
bar(real.(log.(Complex.(e.values))))
bar(real.(e.values))
```

```julia
panel = load_brexit_panel()
s = SimpleSCM(panel)
SynthControl.fit!(s)
```

```julia
numItr = 1000
K = 4
σ = mean(X)
α = 1.0/(I-1) * ones(I-1)
gs = initGibbs(A, I-1, T, K, σ, σ, σ, σ, σ, α)

samplePrior!(gs)
gs.Y = deepcopy(Y)
gs.X = deepcopy(X)

sampleY = zeros(numItr, A, T)

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
end

@showprogress for i ∈ 1:numItr
    updateGibbs!(gs, mask)
    sampleY[i, :, :] = deepcopy(gs.Y)
end
```

```julia
plot([YObs[1, 22, 1:30]; YObs[2, 22, 31:end]], label="Observed")
plot!(mean(sampleY[:, 1, :], dims=1)', label="Bayesian Synthetic")
plot!(s.ŷ₁, label="Abadie Synthetic")
plot!([30], label="Intervention",  seriestype = :vline, line=:dash)
```
